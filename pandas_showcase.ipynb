```python
#5 Create a function that will take a dataframe and a column name. This function should return a series with binary values indicating whether the values
# from the column are above the mean value of that column (indicated with a value of 1 or 0 otherwise). If the value in the column is missing (NaN) 
# the value in the series should also be missing (NaN). Test your function. Hint: search how to check if something is None so that we can return None.

def above_mean_binary(df, column_name):
    """
    Function that returns a series with binary values (1 or 0) based on whether the values
    in the specified column are above the column's mean. NaN values are preserved.
    
    Parameters:
    df (DataFrame): The input pandas DataFrame.
    column_name (str): The name of the column to apply the function to.
    
    Returns:
    Series: A pandas series with binary values indicating above mean (1), below or equal to mean (0), and NaN preserved.
    """
    # Check if the column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in the DataFrame.")
    
    # Calculate the mean of the column, ignoring NaN values
    mean_value = df[column_name].mean()

    # Return a new series with 1 if the value is above the mean, 0 otherwise, and NaN for missing values
    return df[column_name].apply(lambda x: 1 if x > mean_value else (0 if pd.notna(x) else np.nan))

above_mean_binary(data, "Human Capital Index")
```
